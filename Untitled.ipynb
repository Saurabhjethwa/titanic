```python
import pandas as pd

import numpy as np

import seaborn as sns

from sklearn.linear_model import LogisticRegressionCV
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('C:\\Users\\Polestar User\\Documents\\Classification\\Titanic\\train.csv')
test = pd.read_csv('C:\\Users\\Polestar User\\Documents\\Classification\\Titanic\\test.csv')
```

```python
train['Fare'] = train['Fare'].fillna(5)

train['Embarked'] = train['Embarked'].fillna('S') 

train["Cabin"] = train["Cabin"].fillna(0)

train["Cabin"] = np.where(train["Cabin"]==0,0,1)

train['Title'] = train.Name.str.extract(' ([A-Za-z]+)\.', expand=False)


title_dict = {'Capt':0, 'Col':0,
    'Don':0, 'Dr':0, 'Major':0, 'Rev':0, 'Jonkheer':0, 'Dona':0,'Countess':1, 'Lady':1, 'Sir':1,'Mlle':2, 'Miss':2,'Ms':2, 'Miss':2,'Mme':3, 'Mrs':3,"Master":4,"Mr":5}

train['Title'].replace(title_dict,inplace=True)

train['Age'] = train['Age'].fillna(0)

r = [0,16,32,48,64,120]
g = [0,1,2,3,4]
train['Age'] = pd.cut(train['Age'],bins = r,labels=g)

train['Age'] = train['Age'].fillna(0)

p = [0,7.91,14.454,31,520]
q = [0,1,2,3]
train['NewFare'] = pd.cut(train['Fare'],bins = p,labels=q)

train['NewFare']=train['NewFare'].fillna(1)

del train['Fare']

train['Sex'] = np.where(train['Sex']=="male",1,0)

emd_dict = {"S":1,"C":2,"Q":3}
train['Embarked'].replace(emd_dict,inplace = True)

```

```python
del train['PassengerId']

del train['Name']

del train['Ticket']
```
